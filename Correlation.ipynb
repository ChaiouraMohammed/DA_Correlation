```python
import matplotlib.pyplot as plt

targetColumns = ['Zn', 'Li', 'Co', 'Cu', 'Ni', 'Pb', 'V', 'Ba', 'Sr', 'Zr', 'Ce', 'Cr', 'pH ', 'CE']

for i in range(len(targetColumns)):
    for j in range(i+1, len(targetColumns)):
        plt.figure(figsize=(8, 6))
        plt.scatter(data[targetColumns[i]], data[targetColumns[j]], alpha=0.5)
        plt.xlabel(targetColumns[i])
        plt.ylabel(targetColumns[j])
        plt.title(f'Scatter plot of {targetColumns[i]} vs {targetColumns[j]}')
        plt.grid(True)
        plt.show()
```
